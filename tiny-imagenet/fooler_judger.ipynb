```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
import cPickle as pickle

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

%load_ext autoreload
%autoreload 2
```

```python
from cs231n.data_utils import load_tiny_imagenet

tiny_imagenet_a = 'cs231n/datasets/tiny-imagenet-100-A'
        
class_names, X_train, y_train, X_val, y_val, X_test, y_test = load_tiny_imagenet(tiny_imagenet_a)

# Zero-mean the data
mean_img = np.mean(X_train, axis=0)
X_train -= mean_img
X_val -= mean_img
X_test -= mean_img
```

```python
fast_grad_images = pickle.load(open('./data/tiny_foolers_fast/tiny_foolers_fast.p'))
iter_grad_images = pickle.load(open('./data/tiny_foolers/tiny_foolers.p'))
```

```python
fast_grad_images_keys = fast_grad_images.keys()
iter_grad_images_keys = iter_grad_images.keys()
```

```python
X, y = X_val, y_val

num_fast_seen = 0
num_iter_seen = 0
num_clean_seen = 0

num_fast_right = 0
num_iter_right = 0
num_clean_right = 0


while num_fast_seen < 50 or num_iter_seen < 50 or num_clean_seen < 100:
    path = np.random.randint(4)
    if path == 0:
        img = (fast_grad_images[np.random.choice(fast_grad_images_keys)][3][0]+mean_img).transpose(1, 2, 0)
    elif path == 1:
        img = (iter_grad_images[np.random.choice(iter_grad_images_keys)][3][0]+mean_img).transpose(1, 2, 0)
    elif path == 2 or path == 3:
        img = (X[np.random.randint(X.shape[0])] + mean_img).transpose(1, 2, 0)

    plt.imshow(np.round(img).astype('uint8'))
    plt.gca().axis('off')
    plt.gcf().set_size_inches((2, 2))
    plt.show()
    got_answer = False
    while not got_answer:
        answer = raw_input('Are there any visible adversarial artifacts? (y/n)')
        answer = answer.lower()
        if answer not in ['y', 'n']:
            print 'Need "y" or "n" answer'
        else:
            got_answer = True
    
    if path == 0:
        num_fast_seen += 1
        if answer == 'y': num_fast_right += 1
    elif path == 1:
        num_iter_seen += 1
        if answer == 'y': num_iter_right += 1
    elif path == 2 or path == 3:
        num_clean_seen += 1
        if answer == 'n': num_clean_right += 1

    if path == 2 or path == 3: print 'Answer was "n"'
    else: print 'Answer was "y"'
```

```python
print 'Fast Gradient Detection Accuracy', num_fast_right/float(num_fast_seen)
print 'Iterative Adversarial Detection Accuracy', num_iter_right/float(num_iter_seen)
print 'Clean Image Accuracy', num_clean_right/float(num_clean_seen)
```
